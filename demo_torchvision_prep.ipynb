```python
import io
import requests
import torchvision.transforms as T
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_cat_pic = 'https://static01.nyt.com/images/2021/09/14/science/07CAT-STRIPES/07CAT-STRIPES-superJumbo.jpg'
resp = requests.get(url_cat_pic)
img = Image.open(io.BytesIO(resp.content))
```

```python
t = img.size
img2 = img.resize((int(t[0]/5), int(t[1]/5)))
img2
```

```python
preprocess = T.Compose([
   T.Resize(256),
   T.CenterCrop(224),
   T.ToTensor(),
   T.Normalize(
       mean=[0.485, 0.456, 0.406],
       std=[0.229, 0.224, 0.225]
   )
])

x = preprocess(img)
x.shape

# Expected result
# torch.Size([3, 224, 224])
```

```python
reverse_preprocess = T.Compose([
    T.ToPILImage(),
    np.array,
])
```

```python
plt.imshow(reverse_preprocess(x))
```
