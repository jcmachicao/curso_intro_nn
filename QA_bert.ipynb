```python
import pandas as pd
import numpy as np
import os
from transformers import pipeline
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
tokenizer = AutoTokenizer.from_pretrained("mrm8488/bert-base-spanish-wwm-cased-finetuned-spa-squad2-es")
modelo_qa_2 = AutoModelForQuestionAnswering.from_pretrained("mrm8488/bert-base-spanish-wwm-cased-finetuned-spa-squad2-es")
```

```python
preguntas = pipeline('question-answering', model=modelo_qa_2, tokenizer=tokenizer,
                     
                     )
```

```python
contexto = '''
No tardó en darse la convocatoria e intervención de los mejores científicos en infectología y epidemiología de cada país. \
En el caso del Perú, destacan los doctores Elmer Huerta, Eduardo Gotuzzo, Ernesto Bustamante, Willy Lescano, Luis Suárez-Ognio, \
Ciro Maguiña, así como de científicos sociales quienes discutieron, evaluaron, asesoraron y brindaron recomendaciones de la comunidad \
científica sobre la adopción de las mejores decisiones del gobierno con indiscutible liderazgo de los sectores de salud y economía. \
Más aún, son los especialistas llamados a brindar la explicación didáctica respeto a la naturaleza del COVID-19, orientar mediante un lenguaje \
sencillo sobre la práctica cotidiana que nos corresponde cumplir a los ciudadanos, así como al contenido y alcances de las entrevistas y \
noticias periodísticas difundidas por la televisión y radioemisoras de cobertura nacional. La pandemia ha demostrado que los científicos \
pueden ser mediadores decisivos en la divulgación del conocimiento para su directa aplicación en la sociedad.
'''
```

```python
pregs = [
          '¿A qué personas se les reconoce el esfuerzo?',
          '¿Qué efectos finales se identifican en la reflexión?',
          '¿Qué es lo que más valora el autor del texto?',
          '¿Qué se logra comunicar a la sociedad?',
          '¿Qué enfoque comunicacional tiene la reflexión?',
          '¿Qué aspecto de calidad educativa se puede reconocer en la reflexión?',
          '¿Qué preocupación importante se menciona?'
        ]
```

```python
for preg in pregs:
  respuesta = preguntas({'question': preg, 'context': contexto})
  rs, re = respuesta['start'], respuesta['end']
  print(preg, '** ', respuesta['answer'], '** Confiabilidad: ', "%.2f" % (respuesta['score']*100), ' \n', rs, ' ', re, ' ', contexto[rs:re + 30], '...' )
```
