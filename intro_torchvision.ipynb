```python
import io
import requests
import torchvision.transforms as T
from PIL import Image
```

```python
url_pic = 'https://cdn.britannica.com/91/181391-050-1DA18304/cat-toes-paw-number-paws-tiger-tabby.jpg'
resp = requests.get(url_pic)
img = Image.open(io.BytesIO(resp.content))
img.size
```

```python
img2 = img.resize( (250, 400) )
img2
```

```python
preprocess = T.Compose([
   T.Resize(256),
   T.CenterCrop(224),
   T.ToTensor(),
   T.Normalize(
       mean=[0.485, 0.456, 0.406],
       std=[0.229, 0.224, 0.225]
   )
])

x = preprocess(img)
x.shape
```
